```python
df=pd.read_excel(r"D:\DATA\Python\Customer Call List.xlsx")
```

```python
# DROP DUPLICATES
df=df.drop_duplicates()
df
```

```python
#DELETE
df=df.drop(colum='Not_useful_columns')
df
```

```python
# LASTNAME CLEANSING

#Remove ' '  space
#df['Last_Name'].str.strip()
```

```python
#Left/Right/All Strip
#df['Last_Name']=df['Last_Name'].str.lstrip('...')
#df['Last_Name']=df['Last_Name'].str.lstrip('/')
#df['Last_Name']=df['Last_Name'].str.rstrip('_')

df['Last_Name']=df['Last_Name'].str.strip('123._/')
df
```

```python
# Phone Number (eg: keep [^a-zA-Z0-9] only)
#df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)

#Change data type to STRING
#df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))

#Edit phone format xxx-xxx-xxxx
#df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3]+'-'+ x[3:6]+'-'+ x[6:10])

#Remove Na value
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
df
```

```python
#Address Split
df['Address'].str.split(',',n=2, expand=True)
```

```python
df[['Street_Address','State', 'Zipcode']]= df['Address'].str.split(',',n=2, expand=True)
df
```

```python
df=df.drop(columns=['Address','Not_Useful_Column'])
df
```

```python
# Standardize Y/N
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a','N/A')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
#df=df.replace('N/A','')
#df=df.replace('NaN','')

# Fill NaN (NaN is Null value in pandas)
df.fillna('')
df
```

```python
# Drop Customers who don't want to contact
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x, inplace=True)
df
```

```python
# Drop No Phone contact
#Option 1: 
for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x, inplace=True)
df

#Option 2:
df.dropna(subset='Phone_Number', inplace=True)
```
